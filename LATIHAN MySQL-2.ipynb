# One to many relationship

1. Report the account representative for each customer.

```python
run("""
SELECT
    customerNumber,
    lastName,
    salesRepEmployeeNumber
FROM
    customers 
        INNER JOIN
    employees ON salesRepEmployeeNumber = employeeNumber
""")
```

2. Report total payments for Atelier graphique.

```python
run("""
SELECT
    customerName,
    SUM(amount) total
FROM
    customers c
        INNER JOIN
    payments p
    ON c.customerNumber = p.customerNumber
WHERE
    c.customerName ="Atelier graphique"
""")
```

3. Report the total payments by date

```python
run("""
SELECT
    paymentDate,
    SUM(amount) total
FROM
    payments
GROUP BY paymentDate
limit 0,10
""")
```

4. Report the products that have not been sold.

```python
run("""
SELECT
    productCode,
    productName
FROM
    products 
WHERE
    productCode NOT IN (select distinct productCode FROM orderdetails)
""")
```

5. List the amount paid by each customer.

```python
run("""
SELECT
    customerName,
    SUM(amount) as total
FROM 
    customers t1
        INNER JOIN 
    payments t2 ON t1.customerNumber = t2.customerNumber
GROUP BY customerName
""")
```

6. How many orders have been placed by Herkku Gifts?

# Many to many relationship

1. List products sold by order date.

```python
run("""
   SELECT 
       p.productName,
       od.priceEach,
       o.orderDate
   FROM
       products p  INNER JOIN orderdetails od 
   ON 
       p.productCode = od.productCode
       
       INNER JOIN orders  o  
   ON 
       od.orderNumber = o.orderNumber
""")
```

2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
   SELECT 
       p.productName,
       o.orderDate
   FROM
       products p  INNER JOIN orderdetails od 
   ON 
       p.productCode = od.productCode
       
       INNER JOIN orders  o  
   ON 
       od.orderNumber = o.orderNumber
   WHERE
       p.productName = "1940s Ford truck"
   ORDER BY
       o.orderDate desc
""")
```

3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
  SELECT
    c.customerName,
    c.customerNumber, 
    SUM(od.quantityOrdered * od.priceEach) AS Value
FROM customers as c, orders as o, orderdetails as od 
WHERE c.customerNumber = o.customerNumber 
      AND o.orderNumber = od.orderNumber
      GROUP BY o.orderNumber
HAVING Value >25000
limit 0,10    
""")
```

4. Are there any products that appear on all orders?

```python
run("""
SELECT 
    COUNT(orderNumber) jumlahOrder 
FROM orders 
""")
```

5. List the names of products sold at less than 80% of the MSRP.

```python
run("""
SELECT 
    productName, 
    buyPrice, 
    MSRP
FROM 
    products
WHERE 
    buyPrice < (MSRP * 0.8)
    limit 1,15
""")
```

6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
SELECT 
    productName, 
    buyPrice, 
    priceEach
FROM 
    products 
        INNER JOIN 
    orderdetails ON products.productCode = orderdetails.productCode
WHERE priceEach = 2 * buyPrice
    limit 0,16
""")
```

7. List the products ordered on a Monday.

```python
run("""
SELECT 
    p.productname, 
    DAYNAME(o.orderDate) as hari 
FROM 
    products p, orderdetails od, orders o
WHERE 
    p.productCode = od.productCode AND od.orderNumber = o.orderNumber
HAVING 
    hari like "Monday"
limit 0,15 
""")
```

8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT 
    COUNT(p.quantityInStock) as onHoldOrders 
FROM 
    products p, orderdetails od, orders o
WHERE p.productCode = od.productCode AND od.orderNumber = o.orderNumber AND o.status = "On Hold"
""")
```
